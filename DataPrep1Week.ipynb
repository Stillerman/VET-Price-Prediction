```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

#Gathering Data for 1 Week
vet_data_1week = yf.download("VET-USD", start="2021-10-23", end="2021-10-30", interval="1m")

btc_data_1week= yf.download("BTC-USD", start="2021-10-23", end="2021-10-30", interval="1m")
```

```python
for week_hist in [vet_data_1week, btc_data_1week]:
    week_hist['Open_diff'] = week_hist['Open'].diff()
    week_hist['High_diff'] = week_hist['High'].diff()
    week_hist['Close_diff'] = week_hist['Close'].diff()
    week_hist['Volume_diff'] = week_hist['Volume'].diff()
    week_hist['Green'] = week_hist["Open_diff"] > 0
    week_hist['Color'] = ['Green' if g else 'Red' for g in week_hist['Green']]
    
# Join the two df's
week_hist = vet_data_1week.join(btc_data_1week, rsuffix="_BTC", lsuffix="_VET")
len(week_hist[(week_hist["Color_VET"] == "Red") & (week_hist["Color_BTC"] == "Red")])

# Replace the NaN values in the first row with 0's. These NaNs come from the difference function, which requires a previous row
week_hist = week_hist.drop(week_hist.index[0])
week_hist
```

```python
# Drop all rows with NaN values
print(week_hist.shape)
week_hist = week_hist.dropna()
print(week_hist.shape)
```

```python
def backlog(dataset, n):
    og_keys = dataset.keys()
    for key in og_keys:
        for m in range(1, n+1):
            dataset[key + f"_{m}"] = 0
    for i in range(n, len(dataset)):
        for m in range(1, n+1):
            for key in og_keys:
                dataset[key + f"_{m}"] = dataset[key][i-m]

backlog(week_hist, 3)
week_hist.keys()
```

```python
week_hist.to_csv("VET-BTC-Data-1Week.txt")
week_hist.to_csv("VET-BTC-Data-1Week.csv")
print("Done")
```
