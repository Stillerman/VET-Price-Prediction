```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

#Gathering Data for 1 Month
vet_data1 = yf.download("VET-USD", start="2021-10-02", end="2021-10-09", interval="1m")
vet_data2 = yf.download("VET-USD", start="2021-10-09", end="2021-10-16", interval="1m")
vet_data3 = yf.download("VET-USD", start="2021-10-16", end="2021-10-23", interval="1m")
vet_data4 = yf.download("VET-USD", start="2021-10-23", end="2021-10-30", interval="1m")

btc_data1 = yf.download("BTC-USD", start="2021-10-02", end="2021-10-09", interval="1m")
btc_data2 = yf.download("BTC-USD", start="2021-10-09", end="2021-10-16", interval="1m")
btc_data3 = yf.download("BTC-USD", start="2021-10-16", end="2021-10-23", interval="1m")
btc_data4 = yf.download("BTC-USD", start="2021-10-23", end="2021-10-30", interval="1m")

#Concatenate all 4 weeks to make a dataframe of 1 month
vet_data_1month = pd.concat([vet_data1, vet_data2, vet_data3, vet_data4], axis=0)
btc_data_1month = pd.concat([btc_data1, btc_data2, btc_data3, btc_data4], axis=0)
```

```python
for month_hist in [vet_data_1month, btc_data_1month]:
    month_hist['Open_diff'] = month_hist['Open'].diff()
    month_hist['High_diff'] = month_hist['High'].diff()
    month_hist['Close_diff'] = month_hist['Close'].diff()
    month_hist['Volume_diff'] = month_hist['Volume'].diff()
    month_hist['Green'] = month_hist["Open_diff"] > 0
    month_hist['Color'] = ['Green' if g else 'Red' for g in month_hist['Green']]
    
# Join the two df's
month_hist = vet_data_1month.join(btc_data_1month, rsuffix="_BTC", lsuffix="_VET")
len(month_hist[(month_hist["Color_VET"] == "Red") & (month_hist["Color_BTC"] == "Red")])

# Replace the NaN values in the first row with 0's. These NaNs come from the difference function, which requires a previous row
month_hist = month_hist.drop(month_hist.index[0])
month_hist
```

```python
# Drop all rows with NaN values
print(month_hist.shape)
month_hist = month_hist.dropna()
print(month_hist.shape)
```

```python
def backlog(dataset, n):
    og_keys = dataset.keys()
    for key in og_keys:
        for m in range(1, n+1):
            dataset[key + f"_{m}"] = 0
    for i in range(n, len(dataset)):
        for m in range(1, n+1):
            for key in og_keys:
                dataset[key + f"_{m}"] = dataset[key][i-m]

backlog(month_hist, 3)
month_hist.keys()
```

```python
month_hist.to_csv("VET-BTC-Data-1Month.txt")
month_hist.to_csv("VET-BTC-Data-1Month.csv")
print("Done")
```
